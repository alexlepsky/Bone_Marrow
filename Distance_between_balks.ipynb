```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import DBSCAN
from scipy.spatial import distance
import numpy as np
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn.metrics import davies_bouldin_score
import os
```

```python
def get_bm(name):
    try: 
        df_balki = pd.read_table('./ЛХ/Костные балки/' + name +' балки.txt')
        for i in range(len(df_balki)):
            df_balki.iloc[i,0] = float(df_balki.iloc[i,0].replace(',','.'))
            df_balki.iloc[i,1] = float(df_balki.iloc[i,1].replace(',','.'))
    except:
        df_balki = pd.read_table('./ЛХ/Костные балки/' + name +' балки.csv',delimiter = ',',names = ['X','Y']).dropna()
        for i in range(len(df_balki)):
            df_balki.iloc[i,0] = float(df_balki.iloc[i,0])
            df_balki.iloc[i,1] = float(df_balki.iloc[i,1])
        
    try:
        df_megi = pd.read_table('./ЛХ/Мегакариоциты/txt' + name +'.txt')
        for i in range(len(df_megi)):
            df_megi.iloc[i,0] = float(df_megi.iloc[i,0].replace(',','.'))
            df_megi.iloc[i,1] = float(df_megi.iloc[i,1].replace(',','.'))
    except:
        df_megi = pd.read_csv('./ЛХ/Мегакариоциты/csv/' + name +'.csv',delimiter = ',',names = ['X','Y']).dropna()
        for i in range(len(df_megi)):
            df_megi.iloc[i,0] = float(df_megi.iloc[i,0])
            df_megi.iloc[i,1] = float(df_megi.iloc[i,1])
            
    return (df_balki,df_megi)
```

```python
def plotting_bm(b,m):
    b = np.array(b)
    m = np.array(m)
    d = []
    plt.figure(figsize=(50,50))
    plt.axis('equal')
    for i in range(len(m)):
        a = np.array([m[i]])
        c = np.array([b[np.argmin(distance.cdist(b, a, 'euclidean'))]])
        plt.plot([a[0,0],c[0,0]],[a[0,1],c[0,1]],'r',linewidth = 1)
        d.append(np.min(distance.cdist(b, a, 'euclidean')))
    plt.scatter(m[:,0],m[:,1],s=100,c='b')
    plt.scatter(b[:,0],b[:,1],s=30,c='k')
    #plt.savefig('./images/' + i + '.png')
    #plt.close()
    plt.show()
    return np.sort(d)
```

```python
outs = list(outs)
for i in names:
    print(i)
    try:
        b, m = get_bm(i)
        d = []
        b = np.array(b)
        m = np.array(m)
        plt.figure(figsize=(50,50))
        plt.axis('equal')
        for k in range(len(m)):
            a = np.array([m[k]])
            c = np.array([b[np.argmin(distance.cdist(b, a, 'euclidean'))]])
            plt.plot([a[0,0],c[0,0]],[a[0,1],c[0,1]],'r',linewidth = 1)
            d.append(np.min(distance.cdist(b, a, 'euclidean')))
        plt.scatter(m[:,0],m[:,1],s=100,c='b')
        plt.scatter(b[:,0],b[:,1],s=30,c='k')
        plt.title(i)
        plt.show()
        #plt.savefig('./images/' + i + '.png')
        plt.close()
        d = np.sort(d)
        outs.append([i,len(d[d<10])/len(d),len(d[d<50])/len(d),len(d[d<100])/len(d),len(d[d<200])/len(d),len(d[d<500])/len(d),
                   np.mean(d),np.median(d)])
    except:
        print(i,'-----------------')
    
```
